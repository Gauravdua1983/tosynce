```python
#improting necessary libraries
import pandas as pd  #importing pandas
import selenium    #importing selenium
from selenium import webdriver   #importing websriver
from selenium.common.exceptions import NoSuchElementException as e   #importing common exception

#Importing requests
import requests

#importing time
import time
```

```python
#activating the chrome browser
driver=webdriver.Chrome(r'C:\Users\Priyanka\Documents\chromedriver.exe')
```

```python
# Opening amazon.in
driver.get('https://www.amazon.in/')
#maximizing the window
driver.maximize_window()
```

## Image Scraping - Sarees

```python
search_bar = driver.find_element_by_id('twotabsearchtextbox')  # Finding the search bar using it's xpath
image_urls=[]      #empty list to store urls of images

# Finding the search bar using it's xpath
search_bar = driver.find_element_by_id('twotabsearchtextbox')
search_bar.send_keys('Saree')       # Inputing keyword to search images
search_button=driver.find_element_by_id('nav-search-submit-button')   # Finding the xpath of search button
search_button.click()        # Clicking the search button

#fetching the image urls of saree from multiple pages
for j in range(1,7):
    driver.current_url
    image=driver.find_elements_by_xpath('//div[@class="a-section aok-relative s-image-tall-aspect"]/img')
    for img in image:
        image_urls.append(img.get_attribute('src'))
    nxt_button=driver.find_element_by_xpath('//li[@class="a-last"]//a').click()
    time.sleep(5)
```

```python
#checking the length of image urls scraped
len(image_urls)
```

```python
#downloading the images and storing it in a folder locally
for i in range(len(image_urls)):
    print("Downloading {0} of {1} images" .format(i, len(image_urls)))
    response= requests.get(image_urls[i])
    file = open(r"C:\Users\Priyanka\Downloads\amazon_images\Saree\img"+str(i)+".jpg", "wb")
    file.write(response.content)
```
